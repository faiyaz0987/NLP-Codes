```python
import spacy

nlp = spacy.load("en_core_web_sm")

def temporal_extraction(text):
    doc = nlp(text)

    # Time (DATE/TIME)
    times = [(ent.text, ent.label_) for ent in doc.ents if ent.label_ in ["DATE","TIME"]]

    # Events (verbs)
    events = [(t.text, t.lemma_) for t in doc if t.pos_ == "VERB"]

    # Pair them
    pairs = [(e[0], t[0]) for e in events for t in times]

    return times, events, pairs


text = "Apple acquired Beats in 2014. Meeting is tomorrow at 10 AM."

t, e, p = temporal_extraction(text)

print("Times:", t)
print("Events:", e)
print("Pairs:", p)
```
