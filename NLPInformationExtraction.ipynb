```python
!python -m spacy download en_core_web_md
import spacy
from spacy.matcher import Matcher

nlp = spacy.load("en_core_web_md")

# Pattern (Subject → Verb → Object)
pattern = [
    {"DEP": "nsubj"},
    {"DEP": "aux", "OP": "?"},
    {"POS": "VERB"},
    {"OP": "*"},
    {"DEP": {"IN": ["dobj", "attr", "pobj"]}}
]

# Function
def extract_svo(doc):
    matcher = Matcher(nlp.vocab)
    matcher.add("SVO", [pattern])

    relations = []
    for _, start, end in matcher(doc):
        span = doc[start:end]
        subj = span[0].text
        verb = [t.lemma_ for t in span if t.pos_ == "VERB"][0]
        obj = span[-1].text
        relations.append((subj, verb, obj))

    return relations

# Test
text = "Emma visited many places. She saw the Eiffel Tower."
doc = nlp(text)

print("Relations:", extract_svo(doc))

# Named Entity
print("\nEntities:")
for ent in doc.ents:
    print(ent.text, "->", ent.label_)
```
